```python
#Answer the Pandas exercises
# Exercise 1

import pandas as pd
import numpy as np
gameSales = pd.read_csv('dato.csv')
firstParty = pd.read_csv('first_party.csv')

gameSales["NA_Sales"] = gameSales["NA_Sales"].apply(lambda x:  str(x).replace(",","."))
gameSales["NA_Sales"] = gameSales["NA_Sales"].astype(np.float32)
gameSales["EU_Sales"] = gameSales["EU_Sales"].apply(lambda x:  str(x).replace(",","."))
gameSales["EU_Sales"] = gameSales["EU_Sales"].astype(np.float32)
gameSales["JP_Sales"] = gameSales["JP_Sales"].apply(lambda x:  str(x).replace(",","."))
gameSales["JP_Sales"] = gameSales["JP_Sales"].astype(np.float32)
gameSales["Other_Sales"] = gameSales["Other_Sales"].apply(lambda x:  str(x).replace(",","."))
gameSales["Other_Sales"] = gameSales["Other_Sales"].astype(np.float32)
gameSales["Global_Sales"] = gameSales["Global_Sales"].apply(lambda x:  str(x).replace(",","."))
gameSales["Global_Sales"] = gameSales["Global_Sales"].astype(np.float32)

# Ajustes para que coincidan las mayúsculas y las minúsculas
firstParty["Platform"] = firstParty["Platform"].apply(lambda x:  str(x).replace("x","X"))
firstParty["Platform"] = firstParty["Platform"].apply(lambda x:  str(x).replace("XONE","XOne"))
firstParty["Platform"] = firstParty["Platform"].apply(lambda x:  str(x).replace("WIIU","WiiU"))

gameSalesMerge = pd.merge(gameSales,firstParty,how='outer',on=('Platform'), indicator=True)

# Plataformas que no tienen first party:
noFirstPartyPlatforms = gameSalesMerge[gameSalesMerge["First_Party"].isna()]
print("Plataformas sin First Party: ", noFirstPartyPlatforms["Platform"].unique())

# Consola que no tiene ventas:
noSalesEntries = gameSales[["Platform", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].copy()
print("Consolas con al menos un juego sin ventas: ", noSalesEntries[noSalesEntries.isna().any(axis=1)]["Platform"].unique())

print("-----------------------")

# Predecimos datos invalidos de todas las columnas
# Hacemos la media de cada una de las columnas de ventas y con ese valor rellenamos los na
# Mejora: rellenar los nan segun los otros valores de venta

means = {"NA_Sales": 0, "EU_Sales": 0, "JP_Sales": 0, "Other_Sales": 0, "Global_Sales": 0}

for key in means.keys():
    means[key] = gameSales[key].mean()
    print("Media de ", key, " = ", means[key])
    
gameSales = gameSales.fillna(value=means)

# Predecimos el rating buscando la moda de rating del genero del juego
# genres son los generos de los que tenemos que calcular su moda y rellenar solo los NaN de ese genero
genres = gameSales[gameSales.isna().any(axis=1)]["Genre"].unique()

for genre in genres:
    #Estos son los juegos de ese genero
    genreFilter = gameSales[gameSales["Genre"] == genre]
    #Esta es la moda
    modeRating = genreFilter.mode()["Rating"][0]
    
    print("Moda del genero ", genre, " = ", modeRating)
    
    #Relleno con esa moda solo los que cumplan que sean de este genero en la tabla original
    condition = (gameSales["Genre"] == genre)
    gameSales.loc[condition, "Rating"] = gameSales.loc[condition, "Rating"].fillna(modeRating)
    
# Comprobar que no haya NaN
if gameSales[gameSales.isna().any(axis=1)].size == 0:
    print("Ya no quedan NaN en la tabla")
else:
    print("Aun quedan NaN en la tabla")
    display(gameSales[gameSales.isna().any(axis=1)])

print("-----------------------")

# Predecimos datos invalidos de todas las columnas o eliminamos los que no nos interesen a nuestro gusto
videogameDevelopers = pd.read_csv('video-games-developers.csv')

videogameDevelopers["Est."] = videogameDevelopers["Est."].replace('[a-zA-Z.,() ]', '', regex=True, limit=4)
#videogameDevelopers["Est."] = videogameDevelopers["Est."].astype(np.float32)
videogameDevelopers.dropna(subset=["Est."], inplace = True)
videogameDevelopers["Est."] = videogameDevelopers["Est."].astype(np.int32)

#videogameDevelopers["Est."] = pd.to_numeric(videogameDevelopers["Est."], errors='coerce')
#videogameDevelopers.dropna(subset=["Est."], inplace = True)


#Dropeamos las columnas que son mayoritariamente desonocidas y no aportan informacion significatva
videogameDevelopers.drop(columns = ["Administrative division", "Notes"], inplace = True)

#Los desarrolladores que no tengan lanzamientos notables no los queremos
videogameDevelopers.dropna(subset=["Notable games, series or franchises"], inplace = True)

#Los años los estimamos con la media del resto de años
#Hay valores que no tienen el formato correcto. 
print(videogameDevelopers["Est."].mean())

#La ciudad la estimamos con la moda del país

# Comprobar que no haya NaN
if videogameDevelopers[videogameDevelopers.isna().any(axis=1)].size == 0:
    print("Ya no quedan NaN en la tabla")
else:
    print("Aun quedan NaN en la tabla")
    display(videogameDevelopers[videogameDevelopers.isna().any(axis=1)])
```

```python
# Exercise 2
import random
import numpy as np
from scipy.integrate import quad
import time

def integra_mc(fun, a, b, M, num_puntos=10000):
    num_puntos_debajo = 0
    
    for i in range(num_puntos):
        x = random.uniform(a, b)
        y = random.uniform(0, M)
            
        #Compruebo si el punto aleatorio cae debajo de la funcion
        if (y < fun(x)):
            num_puntos_debajo += 1
            
    return num_puntos_debajo / num_puntos * (b-a) * M

def integra_mc_np(fun, a, b, M, num_puntos=10000):
    num_puntos_debajo = 0
    
    x_values = np.random.uniform(a, b, num_puntos)
    y_values = np.random.uniform(0, M, num_puntos)
    
    num_puntos_debajo = np.sum(y_values < fun(x_values))
    
    return num_puntos_debajo / num_puntos * (b - a) * M
    
def sq(x):
    return x * x

start_time = time.time()
result = integra_mc(sq, 0, 10, 100)
end_time = time.time()

print("Sin numpy")
print(f"Resultado: {result}")
print(f"Tiempo: {end_time - start_time} segundos")

start_time = time.time()
result = integra_mc_np(sq, 0, 10, 100)
end_time = time.time()

print("Con numpy")
print(f"Resultado: {result}")
print(f"Tiempo: {end_time - start_time} segundos")
```
